## Let's Deal With Outliers

## To find outliers there are three method ( z score , box plot , domain knowledge )

```python
df1= df.copy()
df1
```

```python
 from scipy import stats
```

```python
z_score = stats.zscore(df1['Age'])
threshold = 3

outliers = (z_score > threshold ) | (z_score < - threshold )
outliers
```

```python
df[outliers == True]
```

```python
outlier_rows = df[outliers]
outlier_rows
```

```python
import seaborn as sns 
```

```python
sns.boxplot(df1['Age'])
```

## lets first seprate the numericals column and non-nuerical column 

```python
numerical_col = df1.select_dtypes(include=['number']).columns
numerical_col
```

```python
numerical_col = df1.select_dtypes(include=['number']).columns.tolist()
categorical_col = df1.select_dtypes(include=['object']).columns.tolist()
print(numerical_col)
print(categorical_col)
```

```python
import matplotlib.pyplot as plt 
```

```python
fig ,axes= plt.subplots(nrows=len(numerical_col),ncols =1 , figsize=(10,6*len(numerical_col)))
for i ,col in enumerate(numerical_col):
    sns.boxplot(x=df1[col] , ax=axes[i])
    z_score = stats.zscore(df1[col])
    threshold = 3
    outlier_indices = (z_score >threshold) | (z_score <-threshold)
    df2=df1[~outlier_indices]
plt.tight_layout()    
plt.show()
```
